```
def display4(a, b, c, title1="Original", title2="Brillo", title3="Ecualizada"):
    plt.figure(figsize=(14,14))
    plt.subplot(231),plt.imshow(a, cmap='gray'), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(232),plt.imshow(b, cmap='gray'), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    img=c
    img=cv2.equalizeHist(img)
    plt.subplot(233),plt.imshow(img, cmap='gray'), plt.title(title3)
    plt.xticks([]), plt.yticks([])
    #plt.xlabel('Intensidad de iluminacion')
    #plt.ylabel('Cantidad de pixeles')
    plt.show()
    
def display2(a, b, c, title1="Original", title2="Oscura", title3="Ecualizada"):
    plt.figure(figsize=(14,14))
    plt.subplot(231),plt.imshow(a, cmap='gray'), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(232),plt.imshow(b, cmap='gray'), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    img=c
    img=cv2.equalizeHist(img)
    plt.subplot(233),plt.imshow(img, cmap='gray'), plt.title(title3)
    plt.xticks([]), plt.yticks([])
    #plt.xlabel('Intensidad de iluminacion')
    #plt.ylabel('Cantidad de pixeles')
    plt.show()    
    
def display3(a, b, c, title1="Histograma Original", title2="Histograma Brillo", title3="Histograma Ecualizado"):
    plt.figure(figsize=(14,14))
    
    img=a
    hist = cv2.calcHist([img], [0], None, [255], [0, 256])
    plt.subplot(231), plt.plot(hist, color='gray'), plt.title(title1)
    plt.xlabel('Intensidad de iluminacion')
    plt.ylabel('Cantidad de pixeles')

    img2=b
    hist2 = cv2.calcHist([img2], [0], None, [255], [0, 256])
    plt.subplot(232), plt.plot(hist2, color='gray'), plt.title(title2)
    plt.xlabel('Intensidad de iluminacion')
    plt.ylabel('Cantidad de pixeles')
    
    img3=c
    img3=cv2.equalizeHist(img3)
    hist3=cv2.calcHist([img3], [0], None, [256], [0,256])
    plt.subplot(233), plt.plot(hist3, color='gray'), plt.title(title3)
    plt.xlabel('Intensidad de iluminacion')
    plt.ylabel('Cantidad de pixeles')
    
    plt.show()
     
def display1(a, b, c, title1="Histograma Original", title2="Histograma Oscuro", title3="Histograma Ecualizado"):
    plt.figure(figsize=(14,14))
    
    img=a
    hist = cv2.calcHist([img], [0], None, [255], [0, 256])
    plt.subplot(231), plt.plot(hist, color='gray'), plt.title(title1)
    plt.xlabel('Intensidad de iluminacion')
    plt.ylabel('Cantidad de pixeles')

    img2=b
    hist2 = cv2.calcHist([img2], [0], None, [255], [0, 256])
    plt.subplot(232), plt.plot(hist2, color='gray'), plt.title(title2)
    plt.xlabel('Intensidad de iluminacion')
    plt.ylabel('Cantidad de pixeles')
    
    img3=c
    img3=cv2.equalizeHist(img3)
    hist3=cv2.calcHist([img3], [0], None, [256], [0,256])
    plt.subplot(233), plt.plot(hist3, color='gray'), plt.title(title3)
    plt.xlabel('Intensidad de iluminacion')
    plt.ylabel('Cantidad de pixeles')
    
    plt.show()    
    
    
```

```
import cv2
import numpy as np

img1 = cv2.imread('scott-pilgrim.jpg', cv2.IMREAD_UNCHANGED)
img_GRAY=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)

img2 = cv2.imread('scott_pilgrim_-_publicity_-_h_2018.jpg', cv2.IMREAD_UNCHANGED)
img_GRAY2=cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)

img3 = cv2.imread('scott-pilgrim-contra-el-mundo-el-videojuego-playstation-store_51583-1.jpg', cv2.IMREAD_UNCHANGED)
img_GRAY3=cv2.cvtColor(img3,cv2.COLOR_BGR2GRAY)

img4 = cv2.imread('02b060ec8fc75c1cad38c9f807913d77.png', cv2.IMREAD_UNCHANGED)
img_GRAY4=cv2.cvtColor(img4,cv2.COLOR_BGR2GRAY)

img = img_GRAY
gamma_two_point_two1=np.array(255*(img/255)**2.2, dtype='uint8')
gamma_point_four1=np.array(255*(img/255)**0.4, dtype='uint8')

img = img_GRAY2
gamma_two_point_two2=np.array(255*(img/255)**2.2, dtype='uint8')
gamma_point_four2=np.array(255*(img/255)**0.4, dtype='uint8')

img = img_GRAY3
gamma_two_point_two3=np.array(255*(img/255)**2.2, dtype='uint8')
gamma_point_four3=np.array(255*(img/255)**0.4, dtype='uint8')

img = img_GRAY4
gamma_two_point_two4=np.array(255*(img/255)**2.2, dtype='uint8')
gamma_point_four4=np.array(255*(img/255)**0.4, dtype='uint8')

```
