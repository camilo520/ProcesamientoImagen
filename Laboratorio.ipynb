```
def displayGray3(a, b, c, title1="Original", title2="Edited", title3="Edited"):
    plt.figure(figsize=(10,10))
    plt.subplot(131),plt.imshow(a, cmap='gray'), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(132),plt.imshow(b, cmap='gray'), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    plt.subplot(133),plt.imshow(c, cmap='gray'), plt.title(title3)
    plt.xticks([]), plt.yticks([])
    plt.show()
    
    
def displayGray2(a, b, title1="Original", title2="Edited"):
    plt.figure(figsize=(10,10))
    plt.subplot(121),plt.imshow(a, cmap='gray'), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(b, cmap='gray'), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    plt.show()
    
    
def displayColor3(a, b, c, title1="Original", title2="Edited", title3="Edited"):
    plt.figure(figsize=(10,10))
    plt.subplot(131),plt.imshow(a), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(132),plt.imshow(b), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    plt.subplot(133),plt.imshow(c), plt.title(title3)
    plt.xticks([]), plt.yticks([])
    plt.show()    
    
    
def displayColor2(a, b, title1="Original", title2="Edited"):
    plt.figure(figsize=(10,10))
    plt.subplot(121),plt.imshow(a), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(b), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    plt.show()        
    
def readRGB(imgPath):
    img=cv2.imread(imgPath)
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return img
  
def displayGray5(a, b, c, d, e, title1="Original", title2="Edited", title3="Edited", title4="Edited", title5="Edited"):
    plt.figure(figsize=(14,14))
    plt.subplot(151),plt.imshow(a, cmap='gray'), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(152),plt.imshow(b, cmap='gray'), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    plt.subplot(153),plt.imshow(c, cmap='gray'), plt.title(title3)
    plt.xticks([]), plt.yticks([])
    plt.subplot(154),plt.imshow(d, cmap='gray'), plt.title(title4)
    plt.xticks([]), plt.yticks([])
    plt.subplot(155),plt.imshow(e, cmap='gray'), plt.title(title5)
    plt.xticks([]), plt.yticks([])
    plt.show()  
    
def displayGray4(a, b, c, d, title1="Original", title2="Edited", title3="Edited", title4="Edited"):
    plt.figure(figsize=(14,14))
    plt.subplot(141),plt.imshow(a, cmap='gray'), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(142),plt.imshow(b, cmap='gray'), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    plt.subplot(143),plt.imshow(c, cmap='gray'), plt.title(title3)
    plt.xticks([]), plt.yticks([])
    plt.subplot(144),plt.imshow(d, cmap='gray'), plt.title(title4)
    plt.xticks([]), plt.yticks([])
    plt.show()      
    
def display4(a, b, c, d, title1="Original", title2="Edited", title3="Edited", title4="Edited"):
    plt.figure(figsize=(14,14))
    plt.subplot(141),plt.imshow(a), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(142),plt.imshow(b, cmap='gray'), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    plt.subplot(143),plt.imshow(c, cmap='gray'), plt.title(title3)
    plt.xticks([]), plt.yticks([])
    plt.subplot(144),plt.imshow(d, cmap='gray'), plt.title(title4)
    plt.xticks([]), plt.yticks([])
    plt.show()    
    
def Threshold(img, thres):
    #img=cv2.imread(img, cv2.IMREAD_GRAYSCALE)
    rows, columns = img.shape
    img2=np.zeros((rows, columns), dtype=np.uint8)
    
    T=thres
    
    for x in range(0,rows):
        for y in range(0, columns):
            r=img[x][y]
            if T<=r:
                img2[x][y]=255
            else:
                img2[x][y]=0
    displayGray2(img, img2)
    return img2   
```

# Punto **1**

```
img=cv2.imread('furi.jpg',0)

gauss=cv2.GaussianBlur(img,(3,3),0)

x_sobel=cv2.Sobel(gauss, cv2.CV_64F, 1, 0, ksize=5)

y_sobel=cv2.Sobel(gauss, cv2.CV_64F, 0, 1, ksize=5)

xy_sobel = x_sobel+y_sobel

displayGray4(gauss, x_sobel, y_sobel, xy_sobel, "Original", "Vertical", "Horizontal", "Diagonal")
```

En esta imagen se puede ver como se pueden detectar bordes verticales incluso en los lapices que esta de manera diagonal, al igual que los bordes horizontales, en los bordes diagonales pasa que solo detecta los de la aprte derecha, en cambio en la parte izquierda no detecta muchos, tal vez por los colores de los lapices que son negros y no se pueden detectar los bordes si no se diferencian los colores.

```
img=cv2.imread('furi2.jpg',0)

gauss=cv2.GaussianBlur(img,(3,3),0)

x_sobel=cv2.Sobel(gauss, cv2.CV_64F, 1, 0, ksize=5)

y_sobel=cv2.Sobel(gauss, cv2.CV_64F, 0, 1, ksize=5)

xy_sobel = x_sobel+y_sobel

displayGray4(img, x_sobel, y_sobel, xy_sobel, "Original", "Vertical", "Horizontal", "Diagonal")
```

Se puede ver que los borde verticales son los que mas abundan en esta imagen, al igual que incluso los bordes diagonales, los borde horizontales son mas bien pocos.

```
img=cv2.imread('furi3.jpg',0)

gauss=cv2.GaussianBlur(img,(3,3),0)

x_sobel=cv2.Sobel(gauss, cv2.CV_64F, 1, 0, ksize=5)

y_sobel=cv2.Sobel(gauss, cv2.CV_64F, 0, 1, ksize=5)

xy_sobel = x_sobel+y_sobel

displayGray4(img, x_sobel, y_sobel, xy_sobel, "Original", "Vertical", "Horizontal", "Diagonal")
```

En esta imagen se puede ver como hay mayor informacion en los bordes diagonales, describen muy bien casi toda la imagen a comparacion de los bordes verticales y horizontales.

Para este punto se uso primero un gaussian blur para reducir el ruido de la imagen, para despues aplicar la funcion sobel de OpenCV, esta funcion se utiliza para calcular las derivadas en X y Y de una imagen, esto lo que hace es que detecta los bordes de una imagen, para detectar los bordes verticales se deriva en X, para detectar los bordes horizontales se deriva en Y, para los bordes diagonales se suman los resultados de las derivadas en X y Y, los resultados se pueden ver en la imagenes.

```
img=cv2.imread('furi.jpg', 0)
img_flotante=np.float32(img)
dft=cv2.dft(img_flotante, flags=cv2.DFT_COMPLEX_OUTPUT)
dft_shift=np.fft.fftshift(dft)
magnitude_spectrum=20*np.log(cv2.magnitude(dft_shift[:,:,0], dft_shift[:,:,1]))
fils, cols = img.shape
centfils, centcols=fils/2, cols/2

mask=np.zeros((fils,cols,2), np.uint8)
mask[np.int32(centfils)-10:np.int32(centfils)+10, np.int32(centcols)-10:np.int32(centcols)+10]=1

fshift=dft_shift*mask
fshift_mask_mag=2000*np.log(cv2.magnitude(fshift[:,:,0], fshift[:,:,1]))

f_shift=np.fft.ifftshift(fshift)
img_back=cv2.idft(f_shift)
img_back=cv2.magnitude(img_back[:,:,0], img_back[:,:,1])
imgfinal=img_back*img

displayGray5(img, img_back, fshift_mask_mag, magnitude_spectrum, imgfinal,"Original", "Filtro pasa bajo", "Mascara", "Espectro", "Con imagen original")

#---------------------------------

centfils, centcols=fils/2, cols/2

mask2=np.ones((fils,cols,2), np.uint8)
mask2[np.int32(centfils)-10:np.int32(centfils)+10, np.int32(centcols)-10:np.int32(centcols)+10]=0

fshift=dft_shift*mask2
fshift_mask_mag2=2000*np.log(cv2.magnitude(fshift[:,:,0], fshift[:,:,1]))

f_shift=np.fft.ifftshift(fshift)
img_back2=cv2.idft(f_shift)
img_back2=cv2.magnitude(img_back2[:,:,0], img_back2[:,:,1])
imgfinal2=img_back2*img

displayGray5(img, img_back2, fshift_mask_mag2, magnitude_spectrum, imgfinal2,"Original", "Filtro pasa alto", "Mascara", "Espectro", "Con imagen original")
```
